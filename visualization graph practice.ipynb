```python
xs = [1,2,3,4,5,6]
ydata = [1,2.8,2.1,4,3,3.5]
ydata2 = [2.2, 1.6, 2.3, 2, 2.3, 2.1]

plt.rcParams.update({'font.size':15, 'font.family':'tahoma'}) #ขนาดและรูปแบบตัวอักษร
plt.figure(figsize=(6,4))

plt.grid(linestyle="--", axis="y")
plt.bar(x, ydata, color='c', width=0.5, label='ในประเทศ') # c=cyan
plt.bar([x+0.2 for x in xs], ydata2, width=0.5, color='orange', alpha=0.8,label="ส่งออก")  #ระยะขยับกราฟแต่ละแท่ง

plt.plot(x, ydata, color='b', marker='s', markersize=8) #b=blue
plt.plot(x, ydata2,color='r', marker='o', markersize=8) #r=red

plt.xlabel('เดือน')
plt.ylabel('ยอดขาย')
plt.title('รายงาน Salse Report')

for tx,ty in list(zip(x, ydata)) : 
    plt.annotate(ty,(tx,ty),textcoords='offset points', xytext=(0,8),ha = 'center',fontsize=12)

plt.legend()
plt.ylim(0, 4.5)
plt.show()
```

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset('tips')
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['size']= df['size'].astype('category')
df.dtypes
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
sns.distplot(df['size'],kde=False)
```

```python
df['size'].value_counts()
```

```python
df.hist('total_bill', bins=20, by='time')
```

```python
df.groupby('time').total_bill.mean().to_frame().sort_values('total_bill',ascending = False)
```

```python
sns.set_style('darkgrid')
grid = sns.FacetGrid(df, row='smoker',col = 'time')
grid.map(plt.hist,'total_bill', bins=10)
```

```python
sns.boxplot(x='time',y='total_bill',data = df, width=0.4, palette = 'husl')
```

```python
df.groupby(['time']).total_bill.quantile([.25, .5, .75]).to_frame().T
```

```python
sns.boxplot(x=df['smoker'], y=df['total_bill'], width=0.5)
```

```python
sns.boxplot(x=df['day'], y=df['total_bill'], width=0.6, hue=df.smoker)
plt.legend(loc='upper left')
```

```python
pvt = df.pivot_table(index=['day'],columns=['smoker'],values=['total_bill'],aggfunc=['sum','count'])
pvt.plot.bar()
pvt
```

```python
sns.violinplot(x=df.time, y=df.total_bill)
```

```python
ord =['Fri','Thur','Sat','Sun']
sns.violinplot(x='day', y='total_bill', data=df, order=ord)
```

```python
sns.violinplot(x='day', y='total_bill', data=df, hue='smoker', order=ord)
```

```python
sns.violinplot(x='day', y='total_bill', data=df, hue='smoker', order=ord, split = True)
```

```python
sns.set_style('whitegrid')
sns.swarmplot(x='sex', y='total_bill', data=df)
```
